##Assignment on whether

```python
def whether():
  data=[0,0]
  while(True):
    tept=int(input("Enter the temperature in degree celisius "))
    hum=int(input("Enter the humidity in percentage "))
    data.append(tept,hum)
    break
  return data
```

```python
def determine(tept,hum):
  if tept>30:
    if hum>60:
      print("Hot Summer ")
    else:
      print("Hot Summer and dry air ")
  elif tept>20:
    if hum>60:
      print("Hot weather ")
    else:
      print("Warm weather with dry air")
  elif tept>10:
      if hum>60:
        print("Cold weather but Humid")
      else:
        print("Cold weather with dry air")
  else:
      if hum>60:
        print("Winter and humid")
      else:
        print("Winter and dry")

def average(datum,ind):
  if not datum:
    return 0
  return sum(r[ind] for r in datum/len(datum))

def main():
  Data=whether()
  if Data==' ':
    print("Data not entered")
    return()

  avegt=average(Data,0)
  avegh=average(Data,1)
  print("Average Tmperature = {avegt:2f} degree celsius")
  print("Average Humidity = {avegh:2f} %")
  d=determine(avegt,avegh)
  print("Weather condition is {d}")





```
