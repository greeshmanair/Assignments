```python
from sklearn.preprocessing import LabelEncoder
el= LabelEncoder()
data['Classification']=el.fit_transform(data['Classification'])
```

```python
data.head()
```

```python
x=data.drop(['Classification'],axis=1)
y=data['Classification']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

###Radial Basis Function

```python
from sklearn.svm import SVC
sv_cls=SVC(kernel='rbf')
sv_cls=sv_cls.fit(x_train,y_train)
```

```python
y_pred_svm=sv_cls.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
accuracy_score(y_test,y_pred_svm)
```

```python
precision_score(y_test,y_pred_svm,average='micro')
```

###Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt1_cls=DecisionTreeClassifier()
dt1_cls=dt1_cls.fit(x_train,y_train)
y_pred_dt1=dt1_cls.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred_dt1)
```

```python
accuracy_score(y_test,y_pred_dt1)
```

```python
precision_score(y_test,y_pred_dt1,average='micro')
```

###Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rd1_cls= RandomForestClassifier()
rd1_cls=rd1_cls.fit(x_train,y_train)
y_pred_rd1=rd1_cls.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred_rd1)
```

```python
accuracy_score(y_test,y_pred_rd1)
```

```python
precision_score(y_test,y_pred_rd1,average='micro')
```

###Linear Regression

```python
sv_cls=SVC(kernel='linear')
sv_cls=sv_cls.fit(x_train,y_train)
```

```python
y_pred_svm=sv_cls.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
accuracy_score(y_test,y_pred_svm)
```

```python
precision_score(y_test,y_pred_svm,average='micro')
```
