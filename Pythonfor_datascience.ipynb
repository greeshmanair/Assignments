###TReating Missing Values

```python
data={'names':['Steve','John','Richard','Sarah','Randy','Michael','Julie'],'age':[20,22,20,21,24,23,22],'gender':['Male','Male','Male','Female','Male','Male','Female'],'rank':[2,1,4,5,3,7,6]}
```

```python
rankings=pd.DataFrame(data)
rankings
```

```python
rankings.iloc[2:5,1]=np.nan
rankings.iloc[3:6,3]=np.nan
rankings.iloc[3,:]=np.nan  #all the columns
```

```python
rankings
```

```python
#handling missing value...
#Detecting missing values
rankings.isnull()
```

```python
rankings.notnull()
```

```python
#Boolean masking
bool_series=pd.isnull(rankings['age'])
rankings[bool_series]
```

```python
#Filling missing values
rankings.fillna(0)
```

```python
#Fill missimg values with prior value
rankings.fillna(method='pad')
```

```python
#Fill missimg values with next value
rankings.fillna(method='bfill')
```

```python
#fill missing values with linear relations....equal distance between values
rankings.interpolate(method='linear')
```

```python
#drop the missing values rows and column
rankings.dropna()
```

```python
#Drop only missing values row
rankings.dropna(how='all')
```

```python
#Drop only missing values column
rankings.dropna(how='all',axis=1)
```

```python
#Drop only missing values row
rankings.dropna(how='all',axis=0)
```

###Handling Duplicates

```python
df=pd.DataFrame({'column 1':[1,1,2,2,3,3,3],'column 2':['a','a','b','b','c','c','c'],'column 3':['A','A','B','B','C','C','C']})
```

```python
df
```

```python
#Detecting duplicates
df.duplicated()
```

```python
#drop all duplicate rows
df.drop_duplicates()
```

```python
#drop all duplicate column
df1=pd.DataFrame({'column 1':[1,1,2,2,3,3,3],'column 2':['a','a','b','b','c','c','c'],'column 3':['A','A','B','B','C','D','C']})
df1.drop_duplicates(['column 3'])
```

###Concaternation of Data and transformation of data

```python
df_obj=pd.DataFrame(np.arange(36).reshape(6,6))
df_obj
```

```python
df_obj2=pd.DataFrame(np.arange(15).reshape(5,3))
df_obj2
```

```python
#concaternation
pd.concat([df_obj,df_obj2],axis=1)  #by row
```

```python
pd.concat([df_obj,df_obj2],axis=0)  #by column
```

```python
#dropping
df_obj.drop([0,2])   #row
```

```python
df_obj.drop([0,2],axis=1)   #column
```

```python
#Adding
series_obj=pd.Series(np.arange(6))
series_obj.name='series added'
series_obj
```

```python
#join method
var_add=pd.DataFrame.join(df_obj,series_obj)
var_add
```
