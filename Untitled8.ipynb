###KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,16):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans=kmeans.fit(data)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,16),wcss)
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#From figure we get k=4
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
Ykmeans=kmeans.fit_predict(data)
```

```python
Ykmeans
```

```python
from sklearn.metrics import silhouette_score
sil_k=silhouette_score(data,Ykmeans)
```

```python
sil_k
```

###Agglomerative Clustering

```python
import scipy.cluster.hierarchy as sch
```

```python
dg=sch.dendrogram(sch.linkage(data,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Data')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ahc=AgglomerativeClustering(n_clusters=2,linkage='ward')
y_ahc=ahc.fit_predict(data)
```

```python
y_ahc
```

```python
sil_ahc=silhouette_score(data,y_ahc)
```

```python
sil_ahc
```
